## When we acquire data for anaylsis then there are possibilities that your data contains missing and duplicate values. For good analysis you need to remove duplicates and fill missing values so that your final result would be precise.
## Missing data can occur when no information is provided for items. these missing values creates vey big problems in real life scenarios.

# Following can be the reasons for missing data:
* Loss of information
* disagreement of data uploading
* Data Unavailability at the time of creation of DataFrame.
* Exists but not collected
* not exists

###  In Pandas missing data represnted by two values
* None 
* NaN

To find missing values in DataFrame there are few functions in Pandas.
* isnull()
* notnull()
* dropna()
* fillna()
* replace()
* interpolate()

isnull() and notnull() are used to check null values in Series as well.

```python
# dropna() with columns

df.dropna(axis = 1)
```

```python
# fillna() filling one value in whole dataframe

df.fillna(0,limit = 1)
```

```python
# filling values according to particular column 

df1 = df.fillna({
    'Gender': 'Male',
    'Team': 'Finance',  
})
```

```python
# fillna()  with forward value
df2 = df.fillna(method = 'ffill')
df2
```
