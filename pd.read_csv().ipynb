# pd.read_csv()
The read_csv() function has an argument called **header** that allows you to specify the headers to use.
If your file has **no header** then you have to tell **read_csv** function by mentioning **header=None**. After this read_csv function automatically create header (numbers) to your file.
You can mention any row as a header by providing **header= index**. This step will make any index a header.
E.g. **header=2** means 2 row will be your header.

```python
df = pd.read_csv('H:\\Projects\\Projects\\Avocado Price  Analysis\\avocado\\avocado.csv')
```

```python
df
```

# usecols
usecols is used to take few columns from the existing DataFrame.

For this header value should be none or default which is not need to mention.

**header=None**, if you want to select columns from the row then you have to mention which row and what values as a column.

You need to pass a list of items to **usecols**.

```python
df1=pd.read_csv('H:\\Projects\\Projects\\Avocado Price  Analysis\\avocado\\avocado.csv',header=0,usecols=['type','Total Volume','year','region'])
```

```python
df1
```

WE can pass column index **to usecols**

usecols=[0,1,2,3,4,5] this will give the columns which are present at the respective index or posotion

```python
df2=pd.read_csv('H:\\Projects\\Projects\\Avocado Price  Analysis\\avocado\\avocado.csv',usecols=[1,3,4,5])
```

```python
df2
```

# Parsing date columns
**Date columns** are represented as ***objects*** by default when loading data from a CSV file.

To read the **date column** correctly, we can use the **argument-parse_dates** to specify a **list of date columns**.

```python
df2.info()
```

Example pd.read_csv('datafile',**parse_dates**=['Date'])

```python
df3 = pd.read_csv('H:\\Projects\\Projects\\Avocado Price  Analysis\\avocado\\avocado.csv',parse_dates=['Date'])
```

```python
df3
```

#### Sometime date is split up into multiple columns.

for example, year, month, and day. To combine them into a datetime, we can pass a nested list to **parse_dates**.

```python
df4 = pd.read_csv('H:\\Projects\\Projects\\Avocado Price  Analysis\\avocado\\avocado.csv',parse_dates=[['year','month','day']])
```
